```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
bins = [0, 25, 40, 60, 100]
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.cut(df['age'],bins = bins, labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
# create columns purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90,
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
# here this type is used to change character into the nummber example here we have assume that the if the frequency of purchase is fortnightly then it is will change the fortnightly into the 14 as we assume in the dict
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all() # it will check the all the values are same or not if all the values are same then it will return true otherwise false
```

```python
df = df.drop(columns=['promo_code_used'], axis=1) # here we are droping the promo_code_used column because it is same as discount_applied column so we can drop it
```
